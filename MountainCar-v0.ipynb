## MountainCar-v0

https://gym.openai.com/envs/MountainCar-v0/  
Episode Termination:  
    The car position is more than 0.5  
    Episode length is greater than 200

```python
from dArwIn import Darwin

import gym
import numpy as np
```

```python
env = gym.make('MountainCar-v0')
obs = env.reset()
print(env.observation_space)
print(env.action_space)
print(obs)
for i in range(10):
    obs, rew, done, _ = env.step(env.action_space.sample())
    print(obs, rew, done)
del env
```

```python
def eval_func(genom):
    env = gym.make('MountainCar-v0')
    rewards = []
    for i in range(100):
        obs = env.reset()
        done = False
        reward = 0
        while not done:
            action = genom.predict(obs)
            obs, rew, done, _ = env.step(np.argmax(action))
            reward += rew
        rewards.append(reward)
    return np.mean(rewards)
```

```python
cfg = {'max_genoms': 60,
       'output_activation': 'softmax',
       'activations': ['relu', 'sigmoid', 'tanh', 'leakyRelu'],
       'standard_hidden_node_activation': 'relu',
       'fitness_treshold': 0,
       'fitness_criterion': 'max',
       'n_inputs': 2,
       'n_outputs': 3,
       'survival_fraction': 1/3,
       'complexity_malus': 0.001,
       'max_stagnation': 15,
       'recurrent': False}
```

```python
darwin = Darwin(cfg)
```

```python
while not darwin.solutionFound and darwin.stagnation < cfg['max_stagnation']: 
    darwin.run(eval_func, runs=1, n_jobs=10, verbose=1)
    darwin.bestGenom.visualize()
```

```python
env = gym.make('MountainCar-v0')
obs = env.reset()
done = False
reward = 0
while not done:
    env.render()
    action = darwin.bestGenom.predict(obs)
    obs, rew, done, _ = env.step(np.argmax(action))
    reward += rew
env.close()
del env
print(reward)
```
